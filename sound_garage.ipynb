```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_gr = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_garage.csv', names = labels, encoding='cp949')
data_gr.head()
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

```python
gr_detail = data_gr['detail']
gr_d_tolist = gr_detail.values.tolist()
gr_d_list = ''
for i in gr_d_tolist:
  gr_d_list += i
gr_d_nouns = okt.nouns(gr_d_list)
gr_d_n = ''
for i in gr_d_nouns:
  gr_d_n += i+' '
gr_d_nouns_count = Counter(gr_d_nouns)
stopwords = set(STOPWORDS)
# stopwords.update(['아파트', '행복'])
```

```python
wd_gr_d = WordCloud(max_font_size=200,background_color="white",mask=circle_mask,font_path='/content/drive/MyDrive/Colab Notebooks/sounds/malgun.ttf',stopwords=stopwords).generate(gr_d_n)
fig = plt.figure()
plt.imshow(wd_gr_d, interpolation='bilinear')
plt.axis('off')
wd_gr_d.to_file(filename='/content/drive/MyDrive/Colab Notebooks/wordcloud_img/wd_gr_d.jpg')
```
