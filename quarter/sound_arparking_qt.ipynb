```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_arp = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_arparking.csv', names = labels, encoding='cp949')
data_arp.head()
```

```python
data_arp.dtypes
data_arp['date'] = pd.to_datetime(data_arp['date'], dayfirst=True, errors='coerce')
```

```python
arp_label = data_arp['date']
# arp_label_du = apt_label[~arp_label.duplicated()]
arp_label_count= Counter(arp_label)
arp_label_count
```

```python
qt01 = (data_arp['date'].dt.month >= 1) & (data_arp['date'].dt.month <=3)
qt02 = (data_arp['date'].dt.month >= 4) & (data_arp['date'].dt.month <=6)
qt03 = (data_arp['date'].dt.month >= 7) & (data_arp['date'].dt.month <=9)
qt04 = (data_arp['date'].dt.month >= 10) & (data_arp['date'].dt.month <=12)
```

```python
arp_1qt_count = Counter(qt01)
arp_2qt_count = Counter(qt02)
arp_3qt_count = Counter(qt03)
arp_4qt_count = Counter(qt04)
qt_count = [arp_1qt_count[True],arp_2qt_count[True],arp_3qt_count[True],arp_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/arp_qt_plot.jpg")
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

```python
arp_question = data_arp[data_arp['type'] == "질의"]
arp_question_1qt = arp_question[qt01]
arp_qd_1qt = arp_question_1qt['detail']

arp_check = data_arp[data_arp['type'] == "확인"]
arp_check_1qt = arp_check[qt01]
arp_cd_1qt = arp_check_1qt['detail']

arp_request = data_arp[data_arp['type'] == "건의"]
arp_request_1qt = arp_request[qt01]
arp_rd_1qt = arp_request_1qt['detail']
```

```python
arp_qd_nouns_1qt = okt.nouns(arp_qd_list_1qt)
arp_qd_n_1qt = ''
for i in arp_qd_nouns_1qt:
  arp_qd_n_1qt += i+' '
```
