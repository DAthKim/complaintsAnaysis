```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_ct = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_cluster.csv', names = labels, encoding='cp949')
data_ct.head()
```

```python
data_ct.dtypes
data_ct['date'] = pd.to_datetime(data_ct['date'], dayfirst=True, errors='coerce')
```

```python
ct_label = data_ct['date']
# ct_label_du = ct_label[~ct_label.duplicated()]
ct_label_count= Counter(ct_label)
ct_label_count
```

```python
qt01 = (data_ct['date'].dt.month >= 1) & (data_ct['date'].dt.month <=3)
qt02 = (data_ct['date'].dt.month >= 4) & (data_ct['date'].dt.month <=6)
qt03 = (data_ct['date'].dt.month >= 7) & (data_ct['date'].dt.month <=9)
qt04 = (data_ct['date'].dt.month >= 10) & (data_ct['date'].dt.month <=12)
```

```python
ct_1qt_count = Counter(qt01)
ct_2qt_count = Counter(qt02)
ct_3qt_count = Counter(qt03)
ct_4qt_count = Counter(qt04)
qt_count = [ct_1qt_count[True],ct_2qt_count[True],ct_3qt_count[True],ct_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/ct_qd_plot.jpg")
```
