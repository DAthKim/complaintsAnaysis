```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_gr = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_garage.csv', names = labels, encoding='cp949')
data_gr.head()
```

```python
data_gr.dtypes
data_gr['date'] = pd.to_datetime(data_gr['date'], dayfirst=True, errors='coerce')
```

```python
gr_label = data_gr['date']
# gr_label_du = gr_label[~gr_label.duplicated()]
gr_label_count= Counter(gr_label)
gr_label_count
```

```python
qt01 = (data_gr['date'].dt.month >= 1) & (data_gr['date'].dt.month <=3)
qt02 = (data_gr['date'].dt.month >= 4) & (data_gr['date'].dt.month <=6)
qt03 = (data_gr['date'].dt.month >= 7) & (data_gr['date'].dt.month <=9)
qt04 = (data_gr['date'].dt.month >= 10) & (data_gr['date'].dt.month <=12)
```

```python
gr_1qt_count = Counter(qt01)
gr_2qt_count = Counter(qt02)
gr_3qt_count = Counter(qt03)
gr_4qt_count = Counter(qt04)
qt_count = [gr_1qt_count[True],gr_2qt_count[True],gr_3qt_count[True],gr_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/gr_qt_plot.jpg")
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

## 1분기

```python
gr_question = data_gr[data_gr['type'] == "질의"]
gr_question_1qt = gr_question[qt01]
gr_qd_1qt = gr_question_1qt['detail']

gr_check = data_gr[data_gr['type'] == "확인"]
gr_check_1qt = gr_check[qt01]
gr_cd_1qt = gr_check_1qt['detail']

gr_request = data_gr[data_gr['type'] == "건의"]
gr_request_1qt = gr_request[qt01]
gr_rd_1qt = gr_request_1qt['detail']
```

질의(=0)

```python
gr_qd_tolist_1qt = gr_qd_1qt.values.tolist()
gr_qd_list_1qt = ''
for i in gr_qd_tolist_1qt:
  gr_qd_list_1qt += i
```

```python
gr_qd_nouns_1qt = okt.nouns(gr_qd_list_1qt)
gr_qd_n_1qt = ''
for i in gr_qd_nouns_1qt:
  gr_qd_n_1qt += i+' '
```

```python
gr_qd_nouns_count_1qt = Counter(gr_qd_nouns_1qt)
gr_qd_nouns_count_1qt
```

```python
wd_gr_1qt_qd = WordCloud(background_color = "white",max_font_size=250,mask=circle_mask,
                          font_path='/content/drive/MyDrive/Colab Notebooks/sounds/malgun.ttf',stopwords=stopwords).generate(gr_qd_n_1qt)
fig = plt.figure()
plt.imshow(wd_gr_1qt_qd, interpolation='bilinear')
plt.axis('off')
wd_gr_1qt_qd.to_file(filename='/content/drive/MyDrive/Colab Notebooks/wordcloud_img/wd_gr_1qt_qd.jpg')
```
