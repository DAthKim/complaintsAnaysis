```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_pk = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_park.csv', names = labels, encoding='cp949')
data_pk.head()
```

```python
data_pk.dtypes
data_pk['date'] = pd.to_datetime(data_pk['date'], dayfirst=True, errors='coerce')
```

```python
pk_label = data_pk['date']
# pk_label_du = pk_label[~pk_label.duplicated()]
pk_label_count= Counter(pk_label)
pk_label_count
```

```python
qt01 = (data_pk['date'].dt.month >= 1) & (data_pk['date'].dt.month <=3)
qt02 = (data_pk['date'].dt.month >= 4) & (data_pk['date'].dt.month <=6)
qt03 = (data_pk['date'].dt.month >= 7) & (data_pk['date'].dt.month <=9)
qt04 = (data_pk['date'].dt.month >= 10) & (data_pk['date'].dt.month <=12)
```

```python
pk_1qt_count = Counter(qt01)
pk_2qt_count = Counter(qt02)
pk_3qt_count = Counter(qt03)
pk_4qt_count = Counter(qt04)
qt_count = [pk_1qt_count[True],pk_2qt_count[True],pk_3qt_count[True],pk_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/pk_qt_plot.jpg")
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

## 1분기

```python
pk_question = data_pk[data_pk['type'] == "질의"]
pk_question_1qt = pk_question[qt01]
pk_qd_1qt = pk_question_1qt['detail']

pk_check = data_pk[data_pk['type'] == "확인"]
pk_check_1qt = pk_check[qt01]
pk_cd_1qt = pk_check_1qt['detail']

pk_request = data_pk[data_pk['type'] == "건의"]
pk_request_1qt = pk_request[qt01]
pk_rd_1qt = pk_request_1qt['detail']
```

질의

```python
pk_qd_tolist_1qt = pk_qd_1qt.values.tolist()
pk_qd_list_1qt = ''
for i in pk_qd_tolist_1qt:
  pk_qd_list_1qt += i
```

```python
pk_qd_nouns_1qt = okt.nouns(pk_qd_list_1qt)
pk_qd_n_1qt = ''
for i in pk_qd_nouns_1qt:
  pk_qd_n_1qt += i+' '
```

```python
pk_qd_nouns_count_1qt = Counter(pk_qd_nouns_1qt)
pk_qd_nouns_count_1qt
```

```python
stopwords = set(STOPWORDS)
# stopwords.update(['모집', '임대', '아파트', '행복'])
```
