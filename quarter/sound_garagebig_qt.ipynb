```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_grb = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_garagebig.csv', names = labels, encoding='cp949')
data_grb.head()
```

```python
data_grb.dtypes
data_grb['date'] = pd.to_datetime(data_grb['date'], dayfirst=True, errors='coerce')
```

```python
grb_label = data_grb['date']
# grb_label_du = grb_label[~grb_label.duplicated()]
grb_label_count= Counter(grb_label)
grb_label_count
```

```python
qt01 = (data_grb['date'].dt.month >= 1) & (data_grb['date'].dt.month <=3)
qt02 = (data_grb['date'].dt.month >= 4) & (data_grb['date'].dt.month <=6)
qt03 = (data_grb['date'].dt.month >= 7) & (data_grb['date'].dt.month <=9)
qt04 = (data_grb['date'].dt.month >= 10) & (data_grb['date'].dt.month <=12)
```

```python
grb_1qt_count = Counter(qt01)
grb_2qt_count = Counter(qt02)
grb_3qt_count = Counter(qt03)
grb_4qt_count = Counter(qt04)
qt_count = [grb_1qt_count[True],grb_2qt_count[True],grb_3qt_count[True],grb_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/grb_qt_plot.jpg")
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

## 1분기

```python
grb_question = data_grb[data_grb['type'] == "질의"]
grb_question_1qt = grb_question[qt01]
grb_qd_1qt = grb_question_1qt['detail']

grb_check = data_grb[data_grb['type'] == "확인"]
grb_check_1qt = grb_check[qt01]
grb_cd_1qt = grb_check_1qt['detail']

grb_request = data_grb[data_grb['type'] == "건의"]
grb_request_1qt = grb_request[qt01]
grb_rd_1qt = grb_request_1qt['detail']
```

질의(=0)

```python
grb_qd_tolist_1qt = grb_qd_1qt.values.tolist()
grb_qd_list_1qt = ''
for i in grb_qd_tolist_1qt:
  grb_qd_list_1qt += i
```

```python
grb_qd_nouns_1qt = okt.nouns(grb_qd_list_1qt)
grb_qd_n_1qt = ''
for i in grb_qd_nouns_1qt:
  grb_qd_n_1qt += i+' '
```

```python
grb_qd_nouns_count_1qt = Counter(grb_qd_nouns_1qt)
grb_qd_nouns_count_1qt
```

```python
stopwords = set(STOPWORDS)
# stopwords.update(['캠핑장','합강','오토'])
```

```python
wd_grb_1qt_qd = WordCloud(background_color = "white",max_font_size=250,mask=circle_mask,
                          font_path='/content/drive/MyDrive/Colab Notebooks/sounds/malgun.ttf',stopwords=stopwords).generate(grb_qd_n_1qt)
fig = plt.figure()
plt.imshow(wd_grb_1qt_qd, interpolation='bilinear')
plt.axis('off')
wd_grb_1qt_qd.to_file(filename='/content/drive/MyDrive/Colab Notebooks/wordcloud_img/wd_grb_1qt_qd.jpg')
```
