```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_jcp = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_jcparking.csv', names = labels, encoding='cp949')
data_jcp.head()
```

```python
data_jcp.dtypes
data_jcp['date'] = pd.to_datetime(data_jcp['date'], dayfirst=True, errors='coerce')
```

```python
jcp_label = data_jcp['date']
# jcp_label_du = jcp_label[~jcp_label.duplicated()]
jcp_label_count= Counter(jcp_label)
jcp_label_count
```

```python
qt01 = (data_jcp['date'].dt.month >= 1) & (data_jcp['date'].dt.month <=3)
qt02 = (data_jcp['date'].dt.month >= 4) & (data_jcp['date'].dt.month <=6)
qt03 = (data_jcp['date'].dt.month >= 7) & (data_jcp['date'].dt.month <=9)
qt04 = (data_jcp['date'].dt.month >= 10) & (data_jcp['date'].dt.month <=12)
```

```python
jcp_1qt_count = Counter(qt01)
jcp_2qt_count = Counter(qt02)
jcp_3qt_count = Counter(qt03)
jcp_4qt_count = Counter(qt04)
qt_count = [jcp_1qt_count[True],jcp_2qt_count[True],jcp_3qt_count[True],jcp_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/jcp_qt_plot.jpg")
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

```python
jcp_qd_nouns_1qt = okt.nouns(jcp_qd_list_1qt)
jcp_qd_n_1qt = ''
for i in jcp_qd_nouns_1qt:
  jcp_qd_n_1qt += i+' '
```
