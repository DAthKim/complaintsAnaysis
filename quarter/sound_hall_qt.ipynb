```python
import pandas as pd
import numpy as np
import seaborn as sns
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime as dt
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_hall = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_hall.csv', names = labels, encoding='cp949')
data_hall.head()
```

```python
data_hall.dtypes
data_hall['date'] = pd.to_datetime(data_hall['date'], dayfirst=True, errors='coerce')
```

```python
hall_label = data_hall['date']
# hall_label_du = hall_label[~hall_label.duplicated()]
hall_label_count= Counter(hall_label)
hall_label_count
```

```python
qt01 = (data_hall['date'].dt.month >= 1) & (data_hall['date'].dt.month <=3)
qt02 = (data_hall['date'].dt.month >= 4) & (data_hall['date'].dt.month <=6)
qt03 = (data_hall['date'].dt.month >= 7) & (data_hall['date'].dt.month <=9)
qt04 = (data_hall['date'].dt.month >= 10) & (data_hall['date'].dt.month <=12)
```

```python
hall_1qt_count = Counter(qt01)
hall_2qt_count = Counter(qt02)
hall_3qt_count = Counter(qt03)
hall_4qt_count = Counter(qt04)
qt_count = [hall_1qt_count[True],hall_2qt_count[True],hall_3qt_count[True],hall_4qt_count[True]]
qt = ["1qt","2qt","3qt","4qt"]
qt_plot = sns.barplot(x=qt, y=qt_count, color="black")
qt_plot.figure.savefig("/content/drive/MyDrive/Colab Notebooks/wordcloud_img/hall_qt_plot.jpg")
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

## 1분기

```python
hall_question = data_hall[data_hall['type'] == "질의"]
hall_question_1qt = hall_question[qt01]
hall_qd_1qt = hall_question_1qt['detail']

hall_check = data_hall[data_hall['type'] == "확인"]
hall_check_1qt = hall_check[qt01]
hall_cd_1qt = hall_check_1qt['detail']

hall_request = data_hall[data_hall['type'] == "건의"]
hall_request_1qt = hall_request[qt01]
hall_rd_1qt = hall_request_1qt['detail']
```

질의

```python
hall_qd_tolist_1qt = hall_qd_1qt.values.tolist()
hall_qd_list_1qt = ''
for i in hall_qd_tolist_1qt:
  hall_qd_list_1qt += i
```

```python
hall_qd_nouns_1qt = okt.nouns(hall_qd_list_1qt)
hall_qd_n_1qt = ''
for i in hall_qd_nouns_1qt:
  hall_qd_n_1qt += i+' '
```

```python
hall_qd_nouns_count_1qt = Counter(hall_qd_nouns_1qt)
hall_qd_nouns_count_1qt
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['회관','지방자치'])
```

```python
wd_hall_1qt_qd = WordCloud(background_color = "white",max_font_size=250,mask=circle_mask,
                          font_path='/content/drive/MyDrive/Colab Notebooks/sounds/malgun.ttf',stopwords=stopwords).generate(hall_qd_n_1qt)
fig = plt.figure()
plt.imshow(wd_hall_1qt_qd, interpolation='bilinear')
plt.axis('off')
wd_hall_1qt_qd.to_file(filename='/content/drive/MyDrive/Colab Notebooks/wordcloud_img/wd_hall_1qt_qd.jpg')
```
