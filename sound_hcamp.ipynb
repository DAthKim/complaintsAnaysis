```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
okt = Okt()
import matplotlib.font_manager as fm
import re
from PIL import Image
from io import BytesIO 
import datetime
from collections import Counter
```

```python
labels = ['place','date','messenger','category','type','due','summary','name','phone','email','detail','receive','address','end','period','address_detail','result','result_detail','sign','etc']
data_hmp = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sounds/sounds_place_hcamp.csv', names = labels, encoding='cp949')
data_hmp.head()
```

```python
circle_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/circle_img.jpg'))
```

```python
hmp_detail = data_hmp['detail']
hmp_d_tolist = hmp_detail.values.tolist()
hmp_d_list = ''
for i in hmp_d_tolist:
  hmp_d_list += i
hmp_d_nouns = okt.nouns(hmp_d_list)
hmp_d_n = ''
for i in hmp_d_nouns:
  hmp_d_n += i+' '
hmp_d_nouns_count = Counter(hmp_d_nouns)
stopwords = set(STOPWORDS)
stopwords.update(['합강','캠핑장'])
```

```python
wd_hmp_d = WordCloud(max_font_size=200,background_color="white",mask=circle_mask,font_path='/content/drive/MyDrive/Colab Notebooks/sounds/malgun.ttf',stopwords=stopwords).generate(hmp_d_n)
fig = plt.figure()
plt.imshow(wd_hmp_d, interpolation='bilinear')
plt.axis('off')
wd_hmp_d.to_file(filename='/content/drive/MyDrive/Colab Notebooks/wordcloud_img/wd_hmp_d.jpg')
```
